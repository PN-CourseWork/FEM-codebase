```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# ============================================================
# Test case 1: u(x,y) = x^3 - x^2 y + y^2 - 1
# ============================================================

u1 = x**3 - x**2*y + y**2 - 1

# Laplacian
lap_u1 = sp.diff(u1, x, 2) + sp.diff(u1, y, 2)

# MMS forcing
q1_tilde = -lap_u1

print("TEST CASE 1")
print("-" * 40)
print("u(x,y) =")
sp.pprint(u1)

print("\nLaplacian Δu =")
sp.pprint(lap_u1)

print("\nManufactured forcing q̃(x,y) = -Δu =")
sp.pprint(q1_tilde)

print("\nDirichlet boundary data f(x,y) = u(x,y)")
sp.pprint(u1)

print("\n\n")

# ============================================================
# Test case 2: u(x,y) = x^2 y^2
# ============================================================

u2 = x**2 * y**2

# Laplacian
lap_u2 = sp.diff(u2, x, 2) + sp.diff(u2, y, 2)

# MMS forcing
q2_tilde = -lap_u2

print("TEST CASE 2")
print("-" * 40)
print("u(x,y) =")
print(u2)

print("\nLaplacian Δu =")
print(lap_u2)

print("\nManufactured forcing q̃(x,y) = -Δu =")
print(q2_tilde)

print("\nDirichlet boundary data f(x,y) = u(x,y)")
print(u2)
```
